```python
import pandas as pd
```

```python
gencode = pd.read_table("/Users/ninaxiong/Downloads/gencode.vM25.annotation.gff3", comment="#",
                        sep = "\t", names = ['seqname', 'source', 'feature', 'start' , 'end', 'score', 'strand', 'frame', 'attribute'])
gencode.head()
```

```python
gencode_genes = gencode[(gencode.feature == "gene")][['seqname', 'start', 'end', 'attribute']].copy().reset_index().drop('index', axis=1)
gencode_genes.head()
```

```python
def gene_info(x):
    # Extract gene names
    g_name = list(filter(lambda x: 'gene_name' in x,  x.split(";")))[0].split("=")[1]
    g_id = list(filter(lambda x: 'ID' in x,  x.split(";")))[0].split("=")[1]
    return (g_name, g_id)
```

```python
gencode_genes["gene_name"], gencode_genes["gene_id"] = zip(*gencode_genes.attribute.apply(lambda x: gene_info(x)))
# gencode_genes["gene_id"] = gencode_genes["gene_id"].str.extract("(.+)\.")
gencode_genes = gencode_genes.set_index("gene_id").drop("attribute", axis=1)
gencode_genes.head()
```

```python
gencode_genes.to_csv("../data/gene_locations.tsv", sep="\t")
```
