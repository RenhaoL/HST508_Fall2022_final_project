```python
from matplotlib import pyplot as plt
from upsetplot import generate_counts,plot
import pandas as pd
import json
```

```python
# load the target dictionary
tg_dict_file = open("../data/tg_dict.json","r")
output = tg_dict_file.read()
tg_dict_file.close()
tg_dict = json.loads(output)
```

```python
tg_dict
```

```python
len(tg_dict)
```

```python
all_TAD = tg_dict.keys()
```

```python
all_TAD_with_gene = [key for key in tg_dict.keys() if len(tg_dict[key]) != 0]
len(all_TAD_with_gene)
```

```python
all_TAD_with_3_more_gene = [key for key in tg_dict.keys() if len(tg_dict[key]) >= 3]
len(all_TAD_with_3_more_gene)
```

```python
from upsetplot import from_contents

TAD_info = from_contents({"total TAD":all_TAD,
                            "TAD with genes":all_TAD_with_gene,
                            "TAD with at least 3 genes": all_TAD_with_3_more_gene})
TAD_info
```

```python
from upsetplot import UpSet
plt = UpSet(TAD_info, subset_size='count', show_counts=True).plot()
```
